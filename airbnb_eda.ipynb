## Python Project EDA & Data Visualization - New York AirBnB Listing 2024 

### Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

### Loading Dataset

```python
df = pd.read_csv('new_york_listings_2024.csv', encoding_errors='ignore')
```

### Data Cleaning

```python
df.isnull().sum()
```

```python
# Dropping all missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
# Dropping duplicate rows
df.duplicated().sum()

df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
# Type casting
df.dtypes

df['id'] = df['id'].astype('int64')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.dtypes
```

## EDA

### Univariate Analysis

```python
# Price Distribution

sns.histplot(data=df, x='price')
plt.show()
```

```python
# Identifying outliers in price column
sns.boxplot(data=df,x='price')
plt.show()
```

```python
df_price = df[df['price'] < 1500]

sns.boxplot(data=df_price, x='price')
plt.show()
```

```python
# Filtered Price Distribution 

sns.histplot(data=df_price, x='price',bins=55)
sns.despine()
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
formatter = ticker.FormatStrFormatter('$%1.0f')
plt.gca().xaxis.set_major_formatter(formatter)
plt.show()
```

### Feature Engineering

```python
# Finding the mean price of beds by neighbour hood group
df_price.groupby(by='neighbourhood_group')['price'].mean()
```

```python
# Adding a price per bed column
df_price['price per bed']= df_price['price']/df['beds']
df_price.head()
```

```python
# Mean price per bed
df_price.groupby(by='neighbourhood_group')['price per bed'].mean()
```

### Bivariate Analysis

```python
# Barplot visualizing the price dependency on neighbourhood by roomtype

sns.barplot(data=df_price,x='neighbourhood_group', y='price', hue='room_type')
sns.despine()
plt.title('Price dependency on neighbourhood by roomtype')
plt.xlabel('Neighbourhood')
plt.ylabel('Price')
formatter = ticker.FormatStrFormatter('$%1.0f')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

```python
# Scatterplot visualizing the number of reviews and relative price 

# Filtering
df_reviews = df_price[df_price['number_of_reviews'] >250 ]

sns.scatterplot(data=df_reviews, x='number_of_reviews', y='price', hue='neighbourhood_group')
sns.despine()
plt.title("Number of reviews per locality and respective prices")
plt.ylabel('Price')
plt.xlabel('Number of reviews')
formatter = ticker.FormatStrFormatter('$%1.0f')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

```python
# 2D density heatmap (KDE) to visualize the relationship between ratings and their prices,
# Focusing on more affordable and higher-rated listings. 

#Filtering
df_filtered = df_price[(df_price['price'] < 500) & (df_price['rating'] >= 4.5)]

sns.kdeplot(
    data=df_filtered,
    x="rating", y="price",
    cmap="viridis", fill=True, thresh=0.05
)
sns.despine()
plt.title("Density Heatmap: Rating vs. Price")
plt.xlabel("Rating")
plt.ylabel("Price")
plt.gca().invert_yaxis() 
formatter = ticker.FormatStrFormatter('$%1.0f')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

```python
# Filtering data

# Filter listings by rating > 4.5 and price < 500
df_filtered = df_price[(df_price['rating'] > 4.5) & (df_price['price'] < 500)]

# Group by neighbourhood and room type, then count
counts = df_filtered.groupby(['neighbourhood_group', 'room_type']).size().reset_index(name='count')

# Total counts per neighbourhood to sort by
neighbourhood_order = counts.groupby('neighbourhood_group')['count'].sum().sort_values(ascending=False).index

counts
```

```python
# Plotting a barplot visualizing the count of Value for Money Listings by neighbourhood and room type
# Condition being (Rating > 4.5 & Price < $500)

sns.barplot(data=counts, x='neighbourhood_group', y='count', hue='room_type',order=neighbourhood_order)
sns.despine()
plt.title('Count of Value-for-Money Listings by Neighbourhood and Room Type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.show()
```

```python
#Geographical Distribution of AirBnb Listing

plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Visualizing correlation between different variables

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```
